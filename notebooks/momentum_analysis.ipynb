```python
# ------------------------------------------------------------
# Football Match Momentum Model - Final corrected version
# Argentina vs France | FIFA World Cup Final 2022
# ------------------------------------------------------------

df = events.copy()
df = df[df["team"].notna()].copy()

# ------------------------------------------------------------
# Coordinate extraction
# StatsBomb pitch: x = 0-120, y = 0-80
# ------------------------------------------------------------

df["x"] = df["location"].apply(lambda loc: loc[0] if isinstance(loc, list) else np.nan)
df["y"] = df["location"].apply(lambda loc: loc[1] if isinstance(loc, list) else np.nan)

df["pass_end_x"] = df["pass_end_location"].apply(lambda loc: loc[0] if isinstance(loc, list) else np.nan)
df["pass_end_y"] = df["pass_end_location"].apply(lambda loc: loc[1] if isinstance(loc, list) else np.nan)

df["carry_end_x"] = df["carry_end_location"].apply(lambda loc: loc[0] if isinstance(loc, list) else np.nan)
df["carry_end_y"] = df["carry_end_location"].apply(lambda loc: loc[1] if isinstance(loc, list) else np.nan)

df["pass_progression"] = df["pass_end_x"] - df["x"]
df["carry_progression"] = df["carry_end_x"] - df["x"]

# Remove penalty shootout events, keep regular time + extra time
df_model = df[df["minute"] <= 120].copy()

goalkeepers = ["Damián Emiliano Martínez", "Hugo Lloris"]


def calculate_momentum(row):
    score = 0

    event_type = row["type"]
    player = row["player"]
    x = row["x"]

    # --------------------------------------------------------
    # 1. Shots
    # --------------------------------------------------------
    if event_type == "Shot":
        xg = row["shot_statsbomb_xg"] if pd.notna(row["shot_statsbomb_xg"]) else 0
        shot_type = row.get("shot_type")

        # Penalties are important, but less dominant than open-play xG
        if shot_type == "Penalty":
            score += xg * 14
        else:
            score += xg * 26
            score += 1.5

        # Extra reward for goals, but controlled
        if row.get("shot_outcome") == "Goal":
            score += 2

    # --------------------------------------------------------
    # 2. Progressive passes
    # --------------------------------------------------------
    elif event_type == "Pass":
        if player not in goalkeepers:
            prog = row["pass_progression"]
            end_x = row["pass_end_x"]
            start_x = row["x"]

            if pd.notna(prog) and pd.notna(end_x) and pd.notna(start_x):

                # Only reward real progressive passes from middle/attacking zones
                if prog >= 10 and start_x >= 40:
                    score += min(prog * 0.05, 2)

                    # Extra value only if the progressive pass reaches advanced zones
                    if end_x >= 80:
                        score += 1

                    if end_x >= 100:
                        score += 2

                # Key passes and assists matter even without huge progression
                if row.get("pass_shot_assist") == True:
                    score += 3

                if row.get("pass_goal_assist") == True:
                    score += 4.5

    # --------------------------------------------------------
    # 3. Progressive carries
    # --------------------------------------------------------
    elif event_type == "Carry":
        prog = row["carry_progression"]
        end_x = row["carry_end_x"]
        start_x = row["x"]

        if pd.notna(prog) and pd.notna(end_x) and pd.notna(start_x):

            # Progressive carry from middle/attacking zones
            if prog >= 10 and start_x >= 40:
                score += min(prog * 0.07, 2.5)

                if end_x >= 80:
                    score += 1

                if end_x >= 100:
                    score += 2

    # --------------------------------------------------------
    # 4. Successful dribbles
    # --------------------------------------------------------
    elif event_type == "Dribble":
        if row.get("dribble_outcome") == "Complete":
            score += 1.5

            if pd.notna(x) and x >= 80:
                score += 1

    # --------------------------------------------------------
    # 5. Recoveries and interceptions
    # --------------------------------------------------------
    elif event_type in ["Ball Recovery", "Interception"]:
        if pd.notna(x) and x >= 70:
            score += 2
        else:
            score += 0.3

    # --------------------------------------------------------
    # 6. High pressure
    # --------------------------------------------------------
    elif event_type == "Pressure":
        if pd.notna(x) and x >= 70:
            score += 0.8

    return score


df_model["momentum_score"] = df_model.apply(calculate_momentum, axis=1)

# ------------------------------------------------------------
# Momentum by minute and team
# ------------------------------------------------------------

momentum = (
    df_model
    .groupby(["minute", "team"])["momentum_score"]
    .sum()
    .reset_index()
)

momentum_wide = momentum.pivot(
    index="minute",
    columns="team",
    values="momentum_score"
).fillna(0)

# Ensure every minute exists
all_minutes = pd.DataFrame(
    index=range(0, int(momentum_wide.index.max()) + 1)
)

momentum_wide = (
    all_minutes
    .join(momentum_wide)
    .fillna(0)
)

# Net momentum: positive = Argentina, negative = France
momentum_wide["net_momentum"] = (
    momentum_wide["Argentina"] - momentum_wide["France"]
)

# Smooth momentum to resemble real match momentum charts
momentum_wide["smooth_momentum"] = (
    momentum_wide["net_momentum"]
    .rolling(window=5, center=True, min_periods=1)
    .mean()
)

# ------------------------------------------------------------
# Quick checks
# ------------------------------------------------------------

print("Total momentum by team:")
print(df_model.groupby("team")["momentum_score"].sum().sort_values(ascending=False))

print("\nTop 20 momentum events:")
display(
    df_model[df_model["momentum_score"] > 0][[
        "minute", "second", "team", "type", "player",
        "x", "pass_end_x", "carry_end_x",
        "pass_progression", "carry_progression",
        "shot_statsbomb_xg", "shot_type", "shot_outcome",
        "momentum_score"
    ]]
    .sort_values("momentum_score", ascending=False)
    .head(20)
)

momentum_wide.head()
```

```python
from statsbombpy import sb
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio

pio.renderers.default = "browser"

# ------------------------------------------------------------
# Load match data
# ------------------------------------------------------------

MATCH_ID = 3869685
events = sb.events(match_id=MATCH_ID)

df = events.copy()
df = df[df["team"].notna()].copy()

# ------------------------------------------------------------
# Coordinates
# ------------------------------------------------------------

df["x"] = df["location"].apply(lambda loc: loc[0] if isinstance(loc, list) else np.nan)
df["y"] = df["location"].apply(lambda loc: loc[1] if isinstance(loc, list) else np.nan)

df["pass_end_x"] = df["pass_end_location"].apply(lambda loc: loc[0] if isinstance(loc, list) else np.nan)
df["carry_end_x"] = df["carry_end_location"].apply(lambda loc: loc[0] if isinstance(loc, list) else np.nan)

df["pass_progression"] = df["pass_end_x"] - df["x"]
df["carry_progression"] = df["carry_end_x"] - df["x"]

df_model = df[df["minute"] <= 120].copy()

goalkeepers = ["Damián Emiliano Martínez", "Hugo Lloris"]

# ------------------------------------------------------------
# Momentum model
# ------------------------------------------------------------

def calculate_momentum(row):
    score = 0
    event_type = row["type"]
    player = row["player"]
    x = row["x"]

    if event_type == "Shot":
        xg = row["shot_statsbomb_xg"] if pd.notna(row["shot_statsbomb_xg"]) else 0
        shot_type = row.get("shot_type")

        if shot_type == "Penalty":
            score += xg * 14
        else:
            score += xg * 26
            score += 1.5

        if row.get("shot_outcome") == "Goal":
            score += 2

    elif event_type == "Pass":
        if player not in goalkeepers:
            prog = row["pass_progression"]
            end_x = row["pass_end_x"]
            start_x = row["x"]

            if pd.notna(prog) and pd.notna(end_x) and pd.notna(start_x):
                if prog >= 10 and start_x >= 40:
                    score += min(prog * 0.05, 2)

                    if end_x >= 80:
                        score += 1

                    if end_x >= 100:
                        score += 2

                if row.get("pass_shot_assist") == True:
                    score += 3

                if row.get("pass_goal_assist") == True:
                    score += 4.5

    elif event_type == "Carry":
        prog = row["carry_progression"]
        end_x = row["carry_end_x"]
        start_x = row["x"]

        if pd.notna(prog) and pd.notna(end_x) and pd.notna(start_x):
            if prog >= 10 and start_x >= 40:
                score += min(prog * 0.07, 2.5)

                if end_x >= 80:
                    score += 1

                if end_x >= 100:
                    score += 2

    elif event_type == "Dribble":
        if row.get("dribble_outcome") == "Complete":
            score += 1.5

            if pd.notna(x) and x >= 80:
                score += 1

    elif event_type in ["Ball Recovery", "Interception"]:
        if pd.notna(x) and x >= 70:
            score += 2
        else:
            score += 0.3

    elif event_type == "Pressure":
        if pd.notna(x) and x >= 70:
            score += 0.8

    return score


df_model["momentum_score"] = df_model.apply(calculate_momentum, axis=1)

# ------------------------------------------------------------
# Momentum by minute
# ------------------------------------------------------------

momentum = (
    df_model
    .groupby(["minute", "team"])["momentum_score"]
    .sum()
    .reset_index()
)

momentum_wide = momentum.pivot(
    index="minute",
    columns="team",
    values="momentum_score"
).fillna(0)

all_minutes = pd.DataFrame(index=range(0, int(momentum_wide.index.max()) + 1))

momentum_wide = all_minutes.join(momentum_wide).fillna(0)

momentum_wide["net_momentum"] = momentum_wide["Argentina"] - momentum_wide["France"]

# ------------------------------------------------------------
# Smooth continuous curve
# ------------------------------------------------------------

plot_df = momentum_wide.copy().reset_index()
plot_df = plot_df.rename(columns={"index": "minute"})

plot_df["momentum_curve"] = (
    plot_df["net_momentum"]
    .rolling(window=17, center=True, min_periods=1)
    .mean()
)

plot_df["argentina_area"] = plot_df["momentum_curve"].where(plot_df["momentum_curve"] >= 0, 0)
plot_df["france_area"] = plot_df["momentum_curve"].where(plot_df["momentum_curve"] < 0, 0)

max_y = plot_df["momentum_curve"].max()
min_y = plot_df["momentum_curve"].min()

# ------------------------------------------------------------
# Goals
# ------------------------------------------------------------

goals = df_model[
    (df_model["type"] == "Shot") &
    (df_model["shot_outcome"] == "Goal") &
    (df_model["minute"] < 120)
][["minute", "team", "player", "shot_type"]].copy()

name_map = {
    "Lionel Andrés Messi Cuccittini": "Messi",
    "Ángel Fabián Di María Hernández": "Di María",
    "Kylian Mbappé Lottin": "Mbappé"
}

goals["player_short"] = goals["player"].replace(name_map)

import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np

# -------------------------------
# Prepare data
# -------------------------------

plot_df = momentum_wide.copy().reset_index()
plot_df = plot_df.rename(columns={"index": "minute"})

plot_df["momentum_curve"] = (
    plot_df["net_momentum"]
    .rolling(window=17, center=True, min_periods=1)
    .mean()
)

plot_df["argentina_area"] = plot_df["momentum_curve"].where(plot_df["momentum_curve"] >= 0, 0)
plot_df["france_area"] = plot_df["momentum_curve"].where(plot_df["momentum_curve"] < 0, 0)

max_y = plot_df["momentum_curve"].max()
min_y = plot_df["momentum_curve"].min()

# Correct goals: remove penalty shootout
goals = df_model[
    (df_model["type"] == "Shot") &
    (df_model["shot_outcome"] == "Goal") &
    (df_model["minute"] < 120)
][["minute", "team", "player", "shot_type"]].copy()

name_map = {
    "Lionel Andrés Messi Cuccittini": "Messi",
    "Ángel Fabián Di María Hernández": "Di María",
    "Kylian Mbappé Lottin": "Mbappé"
}

goals["player_short"] = goals["player"].replace(name_map)
goals["goal_label"] = goals["minute"].astype(str) + "' " + goals["player_short"]

# -------------------------------
# Figure with 2 rows
# -------------------------------

fig = make_subplots(
    rows=2,
    cols=1,
    shared_xaxes=True,
    row_heights=[0.18, 0.82],
    vertical_spacing=0.03
)

# -------------------------------
# Goal timeline
# -------------------------------

for team, color, y_pos in [
    ("Argentina", "#74ACDF", 1),
    ("France", "#142646", -1)
]:
    team_goals = goals[goals["team"] == team].copy()

    # Slight vertical offset for very close goals so labels do not overlap
    if team == "France":
        team_goals["y_pos"] = [y_pos - 0.15 if i % 2 == 0 else y_pos + 0.15 for i in range(len(team_goals))]
    else:
        team_goals["y_pos"] = y_pos

    fig.add_trace(
        go.Scatter(
            x=team_goals["minute"],
            y=team_goals["y_pos"],
            mode="markers+text",
            marker=dict(
                size=12,
                color=color,
                line=dict(color="white", width=2)
            ),
            text=team_goals["goal_label"],
            textposition="top center" if team == "Argentina" else "bottom center",
            textfont=dict(size=11, color=color),
            name=f"{team} goals",
            hovertemplate="<b>%{text}</b><br>" + team + " goal<extra></extra>",
            showlegend=False
        ),
        row=1,
        col=1
    )

# Goal vertical lines
for _, goal in goals.iterrows():
    fig.add_vline(
        x=int(goal["minute"]),
        line_width=1,
        line_dash="dot",
        line_color="rgba(100,100,100,0.22)",
        row=2,
        col=1
    )

# -------------------------------
# Momentum chart
# -------------------------------

fig.add_trace(
    go.Scatter(
        x=plot_df["minute"],
        y=plot_df["argentina_area"],
        mode="lines",
        line=dict(width=0),
        fill="tozeroy",
        fillcolor="rgba(116, 172, 223, 0.42)",
        hoverinfo="skip",
        showlegend=False
    ),
    row=2,
    col=1
)

fig.add_trace(
    go.Scatter(
        x=plot_df["minute"],
        y=plot_df["france_area"],
        mode="lines",
        line=dict(width=0),
        fill="tozeroy",
        fillcolor="rgba(20, 38, 70, 0.55)",
        hoverinfo="skip",
        showlegend=False
    ),
    row=2,
    col=1
)

fig.add_trace(
    go.Scatter(
        x=plot_df["minute"],
        y=plot_df["momentum_curve"],
        mode="lines",
        line=dict(color="rgba(18,18,18,0.95)", width=3),
        name="Net momentum",
        hovertemplate="<b>Minute %{x}</b><br>Net momentum: %{y:.2f}<extra></extra>"
    ),
    row=2,
    col=1
)

fig.add_hline(
    y=0,
    line_width=1.2,
    line_color="rgba(0,0,0,0.65)",
    row=2,
    col=1
)

# -------------------------------
# Layout
# -------------------------------

fig.update_layout(
    title=dict(
        text="<b>Argentina vs France</b><br><span style='font-size:15px;color:#6b7280'>FIFA World Cup Final 2022 | Custom Interactive Match Momentum Model</span>",
        x=0.5,
        xanchor="center",
        y=0.96
    ),
    width=1350,
    height=720,
    plot_bgcolor="white",
    paper_bgcolor="white",
    hovermode="x",
    margin=dict(l=80, r=45, t=120, b=75),
    font=dict(
        family="Arial",
        size=12,
        color="#111827"
    )
)

fig.update_xaxes(
    title_text="Minute",
    range=[0, 125],
    tickmode="array",
    tickvals=[0, 15, 30, 45, 60, 75, 90, 105, 120],
    showgrid=False,
    zeroline=False,
    row=2,
    col=1
)

fig.update_yaxes(
    visible=False,
    range=[-1.8, 1.8],
    row=1,
    col=1
)

fig.update_yaxes(
    title_text="Momentum",
    range=[min_y - 1.5, max_y + 1.5],
    showgrid=True,
    gridcolor="rgba(0,0,0,0.06)",
    zeroline=False,
    row=2,
    col=1
)

# Team labels
fig.add_annotation(
    x=0,
    y=1.35,
    text="<b>ARGENTINA GOALS</b>",
    showarrow=False,
    font=dict(size=12, color="#74ACDF"),
    xanchor="left",
    row=1,
    col=1
)

fig.add_annotation(
    x=0,
    y=-1.35,
    text="<b>FRANCE GOALS</b>",
    showarrow=False,
    font=dict(size=12, color="#142646"),
    xanchor="left",
    row=1,
    col=1
)

fig.add_annotation(
    x=0,
    y=max_y * 0.82,
    text="<b>ARGENTINA MOMENTUM</b>",
    showarrow=False,
    font=dict(size=12, color="#74ACDF"),
    xanchor="left",
    row=2,
    col=1
)

fig.add_annotation(
    x=0,
    y=min_y * 0.82,
    text="<b>FRANCE MOMENTUM</b>",
    showarrow=False,
    font=dict(size=12, color="#142646"),
    xanchor="left",
    row=2,
    col=1
)

fig.add_annotation(
    x=1,
    y=-0.12,
    xref="paper",
    yref="paper",
    text="Data: StatsBomb Open Data | Model: custom weighted attacking threat index",
    showarrow=False,
    font=dict(size=11, color="rgba(80,80,80,0.75)"),
    align="right"
)

fig.show(renderer="browser")
```
