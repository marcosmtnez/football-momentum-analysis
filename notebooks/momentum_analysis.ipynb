# Football Match Momentum Analysis

**Argentina vs France — FIFA World Cup Final 2022**

This notebook builds a custom football match momentum model using StatsBomb Open Data and creates an interactive Plotly visualization.

```python
# ------------------------------------------------------------
# Setup
# ------------------------------------------------------------

import sys
import subprocess
from pathlib import Path

required_packages = {
    "statsbombpy": "statsbombpy",
    "pandas": "pandas",
    "numpy": "numpy",
    "plotly": "plotly"
}

for import_name, package_name in required_packages.items():
    try:
        __import__(import_name)
    except ImportError:
        subprocess.check_call([sys.executable, "-m", "pip", "install", package_name])

from statsbombpy import sb
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots

pd.set_option("display.max_columns", None)
pio.renderers.default = "browser"
```

```python
# ------------------------------------------------------------
# Load match data
# ------------------------------------------------------------

MATCH_ID = 3869685  # Argentina vs France, FIFA World Cup Final 2022

events = sb.events(match_id=MATCH_ID)

df = events.copy()
df = df[df["team"].notna()].copy()

print(f"Events loaded: {df.shape[0]:,}")
df.head()
```

```python
# ------------------------------------------------------------
# Basic exploratory checks
# ------------------------------------------------------------

df["type"].value_counts().head(30)
```

```python
# ------------------------------------------------------------
# Helper functions
# ------------------------------------------------------------

def get_x(location):
    # Extract x-coordinate from StatsBomb location list.
    return location[0] if isinstance(location, list) and len(location) > 0 else np.nan

def get_y(location):
    # Extract y-coordinate from StatsBomb location list.
    return location[1] if isinstance(location, list) and len(location) > 1 else np.nan
```

```python
# ------------------------------------------------------------
# Coordinate extraction
# StatsBomb pitch: x = 0-120, y = 0-80
# ------------------------------------------------------------

df["x"] = df["location"].apply(get_x)
df["y"] = df["location"].apply(get_y)

df["pass_end_x"] = df["pass_end_location"].apply(get_x)
df["pass_end_y"] = df["pass_end_location"].apply(get_y)

df["carry_end_x"] = df["carry_end_location"].apply(get_x)
df["carry_end_y"] = df["carry_end_location"].apply(get_y)

df["pass_progression"] = df["pass_end_x"] - df["x"]
df["carry_progression"] = df["carry_end_x"] - df["x"]

# Keep regular time + extra time. Remove penalty shootout events.
df_model = df[df["minute"] < 120].copy()

print(f"Model events after removing penalty shootout: {df_model.shape[0]:,}")
```

## Momentum model

The model is a custom weighted attacking threat index. It rewards actions that move a team closer to creating danger:

- Shots and expected goals
- Progressive passes
- Progressive carries
- Successful dribbles
- High recoveries and interceptions
- Pressures in advanced areas

Penalty shootout events are removed because they do not represent the flow of the match.

```python
# ------------------------------------------------------------
# Custom momentum scoring model
# ------------------------------------------------------------

goalkeepers = ["Damián Emiliano Martínez", "Hugo Lloris"]

def calculate_momentum(row):
    score = 0.0

    event_type = row["type"]
    player = row["player"]
    x = row["x"]

    # 1. Shots
    if event_type == "Shot":
        xg = row["shot_statsbomb_xg"] if pd.notna(row["shot_statsbomb_xg"]) else 0
        shot_type = row.get("shot_type")

        # Penalties are important, but less representative of open-play dominance
        if shot_type == "Penalty":
            score += xg * 14
        else:
            score += xg * 26
            score += 1.5

        # Controlled bonus for goals
        if row.get("shot_outcome") == "Goal":
            score += 2

    # 2. Progressive passes
    elif event_type == "Pass":
        if player not in goalkeepers:
            prog = row["pass_progression"]
            end_x = row["pass_end_x"]
            start_x = row["x"]

            if pd.notna(prog) and pd.notna(end_x) and pd.notna(start_x):
                # Reward meaningful progression from middle/attacking zones
                if prog >= 10 and start_x >= 40:
                    score += min(prog * 0.05, 2)

                    if end_x >= 80:
                        score += 1

                    if end_x >= 100:
                        score += 2

                # Key passes and assists matter even without huge progression
                if row.get("pass_shot_assist") == True:
                    score += 3

                if row.get("pass_goal_assist") == True:
                    score += 4.5

    # 3. Progressive carries
    elif event_type == "Carry":
        prog = row["carry_progression"]
        end_x = row["carry_end_x"]
        start_x = row["x"]

        if pd.notna(prog) and pd.notna(end_x) and pd.notna(start_x):
            if prog >= 10 and start_x >= 40:
                score += min(prog * 0.07, 2.5)

                if end_x >= 80:
                    score += 1

                if end_x >= 100:
                    score += 2

    # 4. Successful dribbles
    elif event_type == "Dribble":
        if row.get("dribble_outcome") == "Complete":
            score += 1.5

            if pd.notna(x) and x >= 80:
                score += 1

    # 5. Recoveries and interceptions
    elif event_type in ["Ball Recovery", "Interception"]:
        if pd.notna(x) and x >= 70:
            score += 2
        else:
            score += 0.3

    # 6. High pressure
    elif event_type == "Pressure":
        if pd.notna(x) and x >= 70:
            score += 0.8

    return score


df_model["momentum_score"] = df_model.apply(calculate_momentum, axis=1)

print("Total momentum by team:")
display(df_model.groupby("team")["momentum_score"].sum().sort_values(ascending=False))
```

```python
# ------------------------------------------------------------
# Aggregate momentum by minute and team
# ------------------------------------------------------------

momentum = (
    df_model
    .groupby(["minute", "team"])["momentum_score"]
    .sum()
    .reset_index()
)

momentum_wide = momentum.pivot(
    index="minute",
    columns="team",
    values="momentum_score"
).fillna(0)

# Ensure every minute exists from 0 to 119
all_minutes = pd.DataFrame(index=range(0, 120))
momentum_wide = all_minutes.join(momentum_wide).fillna(0)

# Ensure both teams exist as columns
for team in ["Argentina", "France"]:
    if team not in momentum_wide.columns:
        momentum_wide[team] = 0

# Net momentum: positive = Argentina, negative = France
momentum_wide["net_momentum"] = momentum_wide["Argentina"] - momentum_wide["France"]

plot_df = momentum_wide.copy().reset_index().rename(columns={"index": "minute"})

# Smooth continuous curve to resemble broadcast-style momentum charts
plot_df["momentum_curve"] = (
    plot_df["net_momentum"]
    .rolling(window=17, center=True, min_periods=1)
    .mean()
)

plot_df["argentina_area"] = plot_df["momentum_curve"].where(plot_df["momentum_curve"] >= 0, 0)
plot_df["france_area"] = plot_df["momentum_curve"].where(plot_df["momentum_curve"] < 0, 0)

max_y = float(plot_df["momentum_curve"].max())
min_y = float(plot_df["momentum_curve"].min())

plot_df.head()
```

```python
# ------------------------------------------------------------
# Goals
# ------------------------------------------------------------

goals = df_model[
    (df_model["type"] == "Shot") &
    (df_model["shot_outcome"] == "Goal") &
    (df_model["minute"] < 120)
][["minute", "team", "player", "shot_type"]].copy()

name_map = {
    "Lionel Andrés Messi Cuccittini": "Messi",
    "Ángel Fabián Di María Hernández": "Di María",
    "Kylian Mbappé Lottin": "Mbappé"
}

goals["player_short"] = goals["player"].replace(name_map)
goals["goal_label"] = "⚽ " + goals["minute"].astype(str) + "' " + goals["player_short"]

goals.sort_values("minute")
```

```python
# ------------------------------------------------------------
# Interactive momentum chart
# ------------------------------------------------------------

from urllib.parse import quote

ARG_COLOR = "#74ACDF"
FRA_COLOR = "#142646"
ARG_FILL = "rgba(96, 165, 250, 0.45)"
FRA_FILL = "rgba(30, 64, 175, 0.55)"
LINE_COLOR = "rgba(18,18,18,0.95)"
PHASE_COLOR = "rgba(80,80,80,0.38)"
GOAL_LINE_OPACITY = 0.62

# Robust SVG flags, because emoji flags do not always render correctly in Plotly/HTML.
def svg_to_uri(svg):
    return "data:image/svg+xml;utf8," + quote(svg)

ARG_FLAG = svg_to_uri("""
<svg xmlns='http://www.w3.org/2000/svg' width='90' height='60' viewBox='0 0 90 60'>
  <rect width='90' height='20' y='0' fill='#75AADB'/>
  <rect width='90' height='20' y='20' fill='#FFFFFF'/>
  <rect width='90' height='20' y='40' fill='#75AADB'/>
  <circle cx='45' cy='30' r='5.2' fill='#F6B40E'/>
</svg>
""")

FRA_FLAG = svg_to_uri("""
<svg xmlns='http://www.w3.org/2000/svg' width='90' height='60' viewBox='0 0 90 60'>
  <rect width='30' height='60' x='0' fill='#0055A4'/>
  <rect width='30' height='60' x='30' fill='#FFFFFF'/>
  <rect width='30' height='60' x='60' fill='#EF4135'/>
</svg>
""")

fig = go.Figure()

# -------------------------------
# Momentum areas and curve
# -------------------------------

fig.add_trace(
    go.Scatter(
        x=plot_df["minute"],
        y=plot_df["argentina_area"],
        mode="lines",
        line=dict(width=0),
        fill="tozeroy",
        fillcolor=ARG_FILL,
        hoverinfo="skip",
        showlegend=False
    )
)

fig.add_trace(
    go.Scatter(
        x=plot_df["minute"],
        y=plot_df["france_area"],
        mode="lines",
        line=dict(width=0),
        fill="tozeroy",
        fillcolor=FRA_FILL,
        hoverinfo="skip",
        showlegend=False
    )
)

fig.add_trace(
    go.Scatter(
        x=plot_df["minute"],
        y=plot_df["momentum_curve"],
        mode="lines",
        line=dict(color=LINE_COLOR, width=3),
        name="Net momentum",
        hovertemplate="<b>Minute %{x}</b><br>Net momentum: %{y:.2f}<extra></extra>"
    )
)

fig.add_hline(
    y=0,
    line_width=1.2,
    line_color="rgba(0,0,0,0.65)"
)

# -------------------------------
# Match phase markers with hover explanations
# -------------------------------

# Give the plot extra vertical room so event labels live inside the same chart.
y_padding = 2.25
y_low = min_y - y_padding
y_high = max_y + y_padding

phase_markers = [
    {"minute": 45, "label": "HT", "description": "Half-time: end of the first half"},
    {"minute": 90, "label": "FT", "description": "Full-time: end of regular time"},
    {"minute": 105, "label": "ET", "description": "Extra-time half-time"},
    {"minute": 120, "label": "ET", "description": "End of extra time; penalty shootout starts after this point"},
]

for marker in phase_markers:
    minute = marker["minute"]
    label = marker["label"]
    description = marker["description"]

    # Scatter trace instead of only a shape so the line has hover text.
    fig.add_trace(
        go.Scatter(
            x=[minute, minute],
            y=[y_low, y_high],
            mode="lines",
            line=dict(color=PHASE_COLOR, width=1, dash="dash"),
            hovertemplate=(
                f"<b>{label} — {description}</b><br>"
                f"Minute {minute}<extra></extra>"
            ),
            showlegend=False
        )
    )

    fig.add_annotation(
        x=minute,
        y=y_high - 0.18,
        text=label,
        showarrow=False,
        font=dict(size=10, color="rgba(80,80,80,0.88)"),
        bgcolor="rgba(255,255,255,0.88)",
        bordercolor="rgba(0,0,0,0.12)",
        borderwidth=1
    )

# -------------------------------
# Goals integrated into the main chart
# -------------------------------

# The goal labels are now inside the same chart, connected to the momentum curve.
# This avoids the previous look of having two separate charts.

curve_lookup = plot_df.set_index("minute")["momentum_curve"].to_dict()

def nearest_curve_value(minute):
    available_minutes = np.array(list(curve_lookup.keys()))
    nearest_minute = available_minutes[np.argmin(np.abs(available_minutes - minute))]
    return float(curve_lookup[nearest_minute])

goal_offsets = {
    # Manual label positioning keeps close goals readable, especially Mbappé 79' and 80'.
    ("Argentina", 22): {"label_x": 20.0, "label_y": y_high - 0.78},
    ("Argentina", 35): {"label_x": 37.5, "label_y": y_high - 0.78},
    ("Argentina", 107): {"label_x": 105.0, "label_y": y_high - 0.78},
    ("France", 79): {"label_x": 76.5, "label_y": y_low + 0.72},
    ("France", 80): {"label_x": 84.0, "label_y": y_low + 1.25},
    ("France", 117): {"label_x": 116.0, "label_y": y_low + 0.72},
}

for _, goal in goals.iterrows():
    minute = int(goal["minute"])
    team = goal["team"]
    color = ARG_COLOR if team == "Argentina" else FRA_COLOR
    curve_y = nearest_curve_value(minute)

    pos = goal_offsets.get(
        (team, minute),
        {
            "label_x": minute,
            "label_y": y_high - 0.78 if team == "Argentina" else y_low + 0.72
        }
    )
    label_x = pos["label_x"]
    label_y = pos["label_y"]

    # Connector from the actual momentum curve point to the label.
    fig.add_trace(
        go.Scatter(
            x=[minute, label_x],
            y=[curve_y, label_y],
            mode="lines",
            line=dict(color=color, width=1.3, dash="dot"),
            opacity=GOAL_LINE_OPACITY,
            hoverinfo="skip",
            showlegend=False
        )
    )

    # Subtle full-height event line to connect each goal with the whole match timeline.
    fig.add_trace(
        go.Scatter(
            x=[minute, minute],
            y=[y_low, y_high],
            mode="lines",
            line=dict(color=color, width=0.8, dash="dot"),
            opacity=0.22,
            hoverinfo="skip",
            showlegend=False
        )
    )

    # Marker exactly on the momentum curve.
    fig.add_trace(
        go.Scatter(
            x=[minute],
            y=[curve_y],
            mode="markers",
            marker=dict(
                size=12,
                color=color,
                symbol="circle",
                line=dict(color="white", width=2)
            ),
            name=f"{team} goal",
            hovertemplate=(
                f"<b>{goal['goal_label']}</b><br>"
                f"{team} goal<br>"
                f"Momentum at goal: {curve_y:.2f}<extra></extra>"
            ),
            showlegend=False
        )
    )

    # Label bubble inside the same chart.
    fig.add_annotation(
        x=label_x,
        y=label_y,
        text=f"<b>{goal['goal_label']}</b>",
        showarrow=False,
        font=dict(
            size=11,
            color=color,
            family="Segoe UI Emoji, Apple Color Emoji, Noto Color Emoji, Arial"
        ),
        bgcolor="rgba(255,255,255,0.88)",
        bordercolor=color,
        borderwidth=1,
        borderpad=4
    )

# -------------------------------
# Team labels with real SVG flags
# -------------------------------

arg_label_y = max_y * 0.78 if max_y > 0 else 1
fra_label_y = min_y * 0.78 if min_y < 0 else -1

fig.add_layout_image(
    dict(
        source=ARG_FLAG,
        xref="x",
        yref="y",
        x=1.5,
        y=arg_label_y + 0.34,
        sizex=5.0,
        sizey=0.48,
        xanchor="left",
        yanchor="top",
        layer="above"
    )
)

fig.add_annotation(
    x=7.2,
    y=arg_label_y,
    text="<b>ARGENTINA</b>",
    showarrow=False,
    font=dict(size=12, color=ARG_COLOR),
    xanchor="left"
)

fig.add_layout_image(
    dict(
        source=FRA_FLAG,
        xref="x",
        yref="y",
        x=1.5,
        y=fra_label_y + 0.34,
        sizex=5.0,
        sizey=0.48,
        xanchor="left",
        yanchor="top",
        layer="above"
    )
)

fig.add_annotation(
    x=7.2,
    y=fra_label_y,
    text="<b>FRANCE</b>",
    showarrow=False,
    font=dict(size=12, color=FRA_COLOR),
    xanchor="left"
)

# -------------------------------
# Layout
# -------------------------------

fig.update_layout(
    title=dict(
        text="<b>Argentina vs France</b><br><span style='font-size:15px;color:#6b7280'>FIFA World Cup Final 2022 | Custom Interactive Match Momentum Model</span>",
        x=0.5,
        xanchor="center",
        y=0.965
    ),
    width=1350,
    height=680,
    plot_bgcolor="rgba(248,250,252,1)",
    paper_bgcolor="white",
    hovermode="closest",
    margin=dict(l=80, r=45, t=110, b=75),
    font=dict(
        family="Arial",
        size=12,
        color="#111827"
    )
)

fig.update_xaxes(
    title_text="Minute",
    range=[0, 125],
    tickmode="array",
    tickvals=[0, 15, 30, 45, 60, 75, 90, 105, 120],
    showgrid=False,
    zeroline=False
)

fig.update_yaxes(
    title_text="Momentum",
    range=[y_low, y_high],
    showgrid=True,
    gridcolor="rgba(0,0,0,0.06)",
    zeroline=False
)

fig.add_annotation(
    x=1,
    y=-0.14,
    xref="paper",
    yref="paper",
    text="Data: StatsBomb Open Data | Model: custom weighted attacking threat index",
    showarrow=False,
    font=dict(size=11, color="rgba(80,80,80,0.75)"),
    align="right"
)

fig.show(renderer="browser")

```
